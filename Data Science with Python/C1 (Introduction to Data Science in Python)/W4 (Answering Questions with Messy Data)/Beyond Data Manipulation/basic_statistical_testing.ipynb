```python
import numpy as np
import pandas as pd

from scipy import stats # statistical testing
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv("grades.csv")
df.head()
```

```python
print(f"{df.shape[0]} rows and {df.shape[1]} columns")
```

```python
early_finishers = df[pd.to_datetime(df["assignment1_submission"]) < "2016"]
early_finishers.head()
```

```python
late_finishers = df[~df.index.isin(early_finishers.index)]
# considering the data which is in the df but it is NOT (~) in early_finishers
late_finishers.head()
# other possible approach would be to do a left merge, only considering the data that is in df but not in early_finishers 
```

```python
print(early_finishers["assignment1_grade"].mean())
print(late_finishers["assignment1_grade"].mean())
```

```python
ttest_ind(early_finishers["assignment1_grade"], late_finishers["assignment1_grade"])
# 18% of chance that the 2 p√≥pulations are the same
# once it is greater than 5%, we need to consider the null hypothesis
```

```python
print(ttest_ind(early_finishers["assignment2_grade"], late_finishers["assignment2_grade"]))
print(ttest_ind(early_finishers["assignment3_grade"], late_finishers["assignment3_grade"]))
print(ttest_ind(early_finishers["assignment4_grade"], late_finishers["assignment4_grade"]))
print(ttest_ind(early_finishers["assignment5_grade"], late_finishers["assignment5_grade"]))
print(ttest_ind(early_finishers["assignment6_grade"], late_finishers["assignment6_grade"]))
# all above 5%
# we do not have enough evidence to say that the populations differ when it comes to grades
# if we consider other chance similarity (more than 5%), assignment 3 could have been considered statistically significant
# when the pvalue is less than alpha(certain %), there is a significant difference
```

```python
# other methods can be used to analyze similarity instead of p-values

df1 = pd.DataFrame([np.random.random(100) for i in range(100)])
# this is creating 100 lists each one with 100 numbers between 0 and 1

df2 = pd.DataFrame([np.random.random(100) for i in range(100)])
```

```python
# let's consider 10% the critical value

def test_columns(alpha=0.1):
    num_diff = 0 # count the number of similatiries
    for column in df1.columns:
        teststat, pval = ttest_ind(df1[column], df2[column])
        # ttest_ind return a tuple: statistic and the p-value
        if (pval <= alpha):
            print(f"Column {column} is statistically significantly different at the critical_value={alpha}, pval={pval}")
            num_diff += 1
    print(f"The number of differences was {num_diff}, which is {float(num_diff)/len(df1.columns)*100}%")


test_columns()
# since the critical_value is 0.1, about 10% of the columns would be the "same"
```

```python
test_columns(0.05)
# about 5%
```
