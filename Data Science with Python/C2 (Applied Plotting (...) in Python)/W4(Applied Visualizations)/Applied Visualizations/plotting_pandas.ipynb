```python
%matplotlib notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

List of styles of colors available in matplotlib:

```python
plt.style.available
```

Selecting one of them to be shown here:

```python
plt.style.use("seaborn-v0_8-colorblind") # using this color style for the plots
```

Creating a simple dataframe:

```python
np.random.seed(123)

# some random dataframe
df = pd.DataFrame({"A": np.random.randn(365).cumsum(0),
                   "B": np.random.randn(365).cumsum(0) + 20,
                   "C": np.random.randn(365).cumsum(0) - 20},
                   index=pd.date_range("1/1/2017", periods=365))

df.head()
```

Line plot:

```python
df.plot(); # the colors are different from the normal due to the style determined previously
```

Scatterplot:

```python
df.plot("A", "B", kind="scatter");
# plotting a scatterplot with information of columns A and B
```

Scatterplot relating the 3 columns:

```python
df.plot.scatter("A", "C", c="B", s=df["B"], colormap="viridis");
# it is possible to use df.plot.scatter instead of passing "scatter" as a parameter, for example
# c is for the color, which will vary depending on the value of B
# s is for the size, which will be the same of column B
```

```python
ax = df.plot.scatter("A", "C", c="B", s=df["B"], colormap="viridis")
ax.set_aspect("equal") # this "corrects" the x-axis
```

Boxplot:

```python
df.plot.box(); # boxplot
# df.plot(kind="box") would also work
```

Histogram:

```python
df.plot.hist(); # histogram
# df.plot(kind="hist") would also work
```

Kernell Density Estimation (DSE):

```python
df.plot.kde(); # kernell density estimation plot
# useful for visualizing an estimate of a variable's probability density function
# df.plot(kind="kde") would also work
```
